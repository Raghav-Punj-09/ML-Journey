# Problem 1: Swapping without 3rd Variable

```python
""" basic Programing """
a1 = int(input("Enter the first number: "))
b1 = int(input("Enter the second number: "))
print("before swapping : ")
print(f"a = {a1} \nb = {b1}")
a1,b1 = b1, a1
print("after swapping : ")
print(f"a = {a1} \nb = {b1}")
```

# Problem 2: BMI Calculator

```python
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
bmi = weight / (height ** 2)
print(f"Your BMI is: {bmi:.2f}")
```

# Problem 3: Grade Classifier

```python
score_obtained = float(input("Enter the score obtained: "))
if score_obtained >= 90:
    print("Grade: A")
elif score_obtained >= 75:
    print("Grade: B")
elif score_obtained >= 60:
    print("Grade: C")
elif score_obtained >= 50:
    print("Grade: D")
elif score_obtained < 50:
    print("Fail")
```

# Problem 4: Multiplication Table

```python
table_number = int(input("Enter the number for which you want the multiplication table: "))
print(f"Multiplication Table for {table_number}:")
for i in range(1, 11):
    print(f"{table_number} x {i} = {table_number * i}")
```

# Problem 5: counting Vowels in string

```python
input_string = input("Enter a string: ")
VOWELS = "aeiou AEIOU"
vowel_count = 0
for char in input_string:
    if char in VOWELS:
        vowel_count += 1
print(f"The number of vowels in the string is: {vowel_count}")
```

# Problem 6: Highest number in a list

```python
entered_list = [int(x) for x in input("Enter numbers separated by commas: ").split(",")]
entered_list.sort()
print("The max number is: ", entered_list[-1])
```

# Problem 7: FizzBuzz

```python
for i in range(1,51):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

# Problem 8: reverse a string manually

```python
entered_string = input("Enter a string: ")
i = len(entered_string) - 1
rev = ""
while i >= 0:
    rev += entered_string[i]
    i -= 1
print(rev)
```

# Problem 9: Number Guessing Game

```python

import random
import time
random_number = random.randint(1, 100)
user_guess = int(input("Guess the number between 1 and 100: "))
while True:
    user_guess = int(input("Again Guess the number between 1 and 100: "))
    print(f"Checking {time.sleep(1)}.{time.sleep(1)}.{time.sleep(1)}.")
    time.sleep(1)
    if user_guess < random_number:
        print("low! Try again.")
        time.sleep(1)
        print(f"Setting yp again, Wait{time.sleep(0.5)}.{time.sleep(0.5)}.{time.sleep(0.5)}.")
    elif user_guess > random_number:
        print("high! Try again.")
        time.sleep(1)
        print(f"Setting yp again, Wait{time.sleep(0.5)}.{time.sleep(0.5)}.{time.sleep(0.5)}.")
    else:
        print("Congratulations! You've guessed the correct number.")
        break
```
